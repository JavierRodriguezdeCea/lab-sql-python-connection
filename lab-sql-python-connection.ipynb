```python
def rentals_month(engine, month: int, year: int):
    with engine.connect() as connection:
        query = text(f'SELECT * FROM RENTAL WHERE MONTH(RENTAL.RENTAL_DATE) = {month} AND YEAR(RENTAL.RENTAL_DATE) = {year}')
        result = connection.execute(query)
        tabla = pd.DataFrame(result.all())
        return tabla
```

```python
rentals_month(engine, 5, 2005)
```

```python
def rental_count_month (engine, month: int, year: int):
    with engine.connect() as connection:
        query = text(f'SELECT RENTAL.CUSTOMER_ID, COUNT(*) AS rentals_{month}_{year} FROM RENTAL WHERE MONTH(RENTAL.RENTAL_DATE) = {month} OR YEAR(RENTAL.RENTAL_DATE) = {year} GROUP BY RENTAL.CUSTOMER_ID')
        result = connection.execute(query)
        tabla1 = pd.DataFrame(result.all())
        return tabla1
```
